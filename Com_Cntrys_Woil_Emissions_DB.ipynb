```python
import pandas as pd
```

```python
world_oil_cntry ="worldoildata_allcountries.csv"
world_oil_cntry = pd.read_csv(world_oil_cntry)
world_oil_cntry

world_oil_cntry = world_oil_cntry.rename(columns = {"Unnamed: 0": "Country"}) 
del world_oil_cntry["REGION_NAME"]
world_oil_cntry

selected_cntry ="selected_cntrys.csv"
selected_cntry = pd.read_csv(selected_cntry)
selected_cntry

selected_cntry = selected_cntry.rename(columns = {"0": "Country"}) 
del selected_cntry["Unnamed: 0"]
selected_cntry

df = pd.merge(world_oil_cntry, selected_cntry, how = 'inner', on = 'Country')
df
```

```python
df.to_csv("common_cntry_Woil_emissions_data.csv", index= True, header = True)
```
